```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
data=pd.read_csv("HistoricalQuotes.csv")
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['volume'],label="Mama yangu")
plt.title("Historical dataset")
plt.xlabel("Dates")
plt.ylabel("prices")
plt.legend()
plt.show()
```

```python
data.plot(subplots=True)
plt.show()
```

```python
data.head()
```

```python
data.columns=data.columns.str.strip()
```

```python
dollar_col=["Close/Last","Open","High","Low"]
for col in dollar_col:
    data[col]=data[col].replace('[\$,]', '',regex=True).astype(float)
```

```python
data.info()
```

```python
data.sort_values('Date',inplace=True)
```

```python
data.Date=pd.to_datetime(data['Date'])
data.set_index('Date',inplace=True)
```

```python
data.plot(subplots=True)
plt.show()
```

```python
close=data["Close/Last"]
```

```python
plt.figure(figsize=(10,5))
plt.plot(close,label="Data columns")
plt.title("Historical dataset")
plt.xlabel("Date")
plt.ylabel("price over time")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Volume'],label="Data columns")
plt.title("Historical dataset")
plt.xlabel("Date")
plt.ylabel("price over time")
plt.legend()
plt.show()
```
