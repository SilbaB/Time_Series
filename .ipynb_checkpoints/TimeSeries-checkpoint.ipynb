Checking for stationarity


```python
def adf_test(close_series):
    result = adfuller(close_series.dropna())
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    if result[1] <= 0.05:
        print("The series is stationary (reject null hypothesis)")
    else:
        print("The series is not stationary (fail to reject null hypothesis)")

adf_test(data['Close/Last'])
```

Data differentiation

```python

Close_Diff=data["Close/Last"].diff().dropna()

```

```python
adf_test_diff = adfuller(Close_Diff.dropna())

print("ADF Statistic (Differenced):", adf_test_diff[0])
print("p-value:", adf_test_diff[1])

if adf_test_diff[1] < 0.05:
    print("The differenced series is stationary.")
else:
    print("Further transformations may be needed.")
```

```python
from statsmodels.tsa.stattools import acf, pacf 
import numpy as np
import matplotlib.pyplot as plt

# Calculate ACF and PACF
lag_acf = acf(Close_Diff.dropna(), nlags=20)
lag_pacf = pacf(Close_Diff.dropna(), nlags=20, method="ols")

# Plot ACF and PACF
plt.figure(figsize=(12, 6))

# ACF plot
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(Close_Diff.dropna())), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(Close_Diff.dropna())), linestyle='--', color='gray')
plt.title('Autocorrelation Function (ACF)')
plt.xlabel('Lag')
plt.ylabel('Correlation')

# PACF plot
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(Close_Diff.dropna())), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(Close_Diff.dropna())), linestyle='--', color='gray')
plt.title('Partial Autocorrelation Function (PACF)')
plt.xlabel('Lag')
plt.ylabel('Partial Correlation')

plt.tight_layout()
plt.show()
```
